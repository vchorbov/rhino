```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set_style("whitegrid")
sns.set_palette("GnBu_d")
import folium
from folium import plugins
from folium.plugins import HeatMap
from datetime import datetime
from datetime import timedelta
import math
import random 
import timeit
```

```python
BRM = pd.read_csv("resources/bremen.csv")
FRB = pd.read_csv("resources/freiburg.csv") 
```

```python
#create the map
FRB_heat_map = folium.Map(location=[48.3669, 7.1485],
                          tiles='Stamen Toner',
                           zoom_start = 13) 


# Ensure you're handing it floats
FRB['orig_lat'] = FRB['orig_lat'].astype(float)
FRB['orig_lng'] = FRB['orig_lng'].astype(float)

 # Reducing data size so it runs faster
FRB_summer = FRB[["day", "time", "b_number","orig_lat", "orig_lng"]]

 # selecting rows based on condition:Season == summer
FRB_summer =  FRB_summer[(FRB_summer['day'] >= "2019-06-01") & 
          (FRB_summer['day'] <= "2019-09-31")] 

# Create weight column, using time as weight
FRB_summer['Weight'] =  pd.to_timedelta(FRB_summer['time']).dt.seconds
FRB_summer['Weight'] = FRB_summer['Weight'].astype(float)


 # List comprehension to make out list of lists
#(0,13)??
heat_data = [[[row['orig_lat'],row['orig_lng']] for index,
              row in FRB_summer[FRB_summer['Weight'] == i].iterrows()]
             for i in range(len(FRB_summer))]

# Plot it on the map
hm = plugins.HeatMapWithTime(heat_data,auto_play=True,max_opacity=0.8)
hm.add_to(FRB_heat_map)

# Display the map
FRB_heat_map

```
