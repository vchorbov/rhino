```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set_style("whitegrid")
sns.set_palette("GnBu_d")
import folium
from folium import plugins
from folium.plugins import HeatMap
from datetime import datetime
from datetime import timedelta
import math
import random 
```

```python
BRM = pd.read_csv("bremen.csv")
FRB = pd.read_csv("freiburg.csv")
```

```python
#avaiability of bikes over time per Day

BRM_Date_avail = BRM.groupby("day")["b_number"].nunique()

fig, ax = plt.subplots(figsize=(12,4), dpi= 80)

ax.plot(BRM_Date_avail)
ax.set_title("Bremens bikes avaiable per day")
plt.show()
```

```python
FRB_Date_avail = FRB.groupby("day")["b_number"].nunique()

fig, ax = plt.subplots(figsize=(12,4), dpi= 80)

ax.plot(FRB_Date_avail)
ax.set_title("Freiburgs bikes avaiable per day")
plt.show()
```

```python
# Avaiability by time of day (Bremen)

BRM_time_avail = BRM.groupby(["day","time"])["b_number"].nunique()
BRM_time_avail = pd.DataFrame(BRM_time_avail)
```

```python
fig,ax = plt.subplots(figsize=(10,4)) 

sns.boxplot(x=BRM_time_avail.index.get_level_values(1), y=BRM_time_avail["b_number"],ax=ax)
plt.show()
```

```python
# Avaiability by time of day (Freiburg)

FRB_time_avail = FRB.groupby(["day","time"])["b_number"].nunique()
FRB_time_avail = pd.DataFrame(FRB_time_avail)
```

```python
fig,ax = plt.subplots(figsize=(10,4)) 

sns.boxplot(x=FRB_time_avail.index.get_level_values(1), y=FRB_time_avail["b_number"],ax=ax)
plt.show()
```
